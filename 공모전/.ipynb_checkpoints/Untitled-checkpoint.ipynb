```python
result_df = pd.read_csv(r'C:\Users\CHOI\공모전\데이터\result.csv', encoding = 'cp949')
highway_df = pd.read_csv(r'C:\Users\CHOI\공모전\데이터\도로정보.csv', encoding = 'cp949')
geongi = pd.read_csv(r'C:\Users\CHOI\공모전\데이터\경기충청_roadkill.csv')
```

```python
# 노선명을 기준으로 그룹화하고 평균 차로 수를 계산하여 새로운 컬럼에 저장
highway_df['노선 평균 차로 수'] = highway_df.groupby('노선번호')['차로수'].transform('mean')

# 쉼표 제거 및 int로 변환
highway_df['계'] = highway_df['계'].str.replace(',', '').astype(int)

# 노선명을 기준으로 그룹화하고 평균 차량 이동량 계산하여 새로운 컬럼에 저장
highway_df['노선 일 평균 이동 차량 수'] = highway_df.groupby('노선번호')['계'].transform('sum')

# 노선명을 기준으로 그룹화하고 평균 차량 이동량 계산하여 새로운 컬럼에 저장
highway_df['노선 길이'] = highway_df.groupby('노선번호')['연장(km)'].transform('sum')

highway_df = highway_df.rename(columns={'노선번호': ' 노선명 '})
```

```python
# highway_df에서 필요한 컬럼 선택
highway_df_selected = highway_df[[' 노선명 ', '노선 평균 차로 수', '노선 길이', '노선 일 평균 이동 차량 수']]

# 노선명 컬럼을 기준으로 중복 제거
highway_df_selected_unique = highway_df_selected.drop_duplicates(subset=[' 노선명 '])

# 선택한 컬럼들을 기준으로 left join 수행
result_df = pd.merge(result_df, highway_df_selected_unique, on=' 노선명 ', how='left')
```

```python
mean_line_count = highway_df['노선 평균 차로 수'].mean()
mean_line_length = highway_df['노선 길이'].mean()
mean_line_car = highway_df['노선 일 평균 이동 차량 수'].mean()
```

```python
# 결측치 보간
result_df['노선 평균 차로 수'].fillna(mean_line_count, inplace=True)
result_df['노선 길이'].fillna(mean_line_length, inplace=True)
result_df['노선 일 평균 이동 차량 수'].fillna(mean_line_car, inplace=True)

# 필요없는 컬럼 드랍
columns_to_drop = [' 본부명 ', ' 지사명 ',' 구간 ', ' 방 향 ', ' 5km ']
result_df = result_df.drop(columns=columns_to_drop, axis=1)

# '반경 동물 개체 수' 컬럼의 0 값을 평균값으로 대체
mean_value = result_df[result_df['반경 동물 개체 수'] != 0]['반경 동물 개체 수'].mean()
result_df['반경 동물 개체 수'] = result_df['반경 동물 개체 수'].replace(0, mean_value)
```

```python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# 초기 파라미터 리스트
columns_to_check = [' 발생건수 ', '울타리 평균 높이',
                    '반경 생태통로 수', '반경 동물 개체 수', 
                    '노선 평균 차로 수', '노선 일 평균 이동 차량 수', 
                    'km 당 차량 수', '개체 마리 당 생태통로', 
                    '개체 마리 당 울타리 길이']



def calculate_vif(data):
    vif_data = pd.DataFrame()
    vif_data["feature"] = data.columns
    vif_data["VIF"] = [variance_inflation_factor(data.values, i) for i in range(len(data.columns))]
    return vif_data

# VIF 계산 함수
def select_features(columns_to_check):
    selected_features = columns_to_check.copy()

    while True:
        vif_data = calculate_vif(result_df[selected_features])
        max_vif_feature = vif_data.iloc[vif_data['VIF'].idxmax()]['feature']
        max_vif_value = vif_data.iloc[vif_data['VIF'].idxmax()]['VIF']

        if max_vif_value <= 10:
            break
        else:
            selected_features.remove(max_vif_feature)

    return selected_features

selected_features = select_features(columns_to_check)
print("Selected Features with VIF <= 10:")
print(selected_features)
```

### 가중치를 고려한 클러스터링

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np
import folium
from kneed import KneeLocator
```

## 클러스터 수 결정

### 엘보우

```python
# 데이터 준비
data = final_df[['위도', '경도']].values
weights = final_df['weight'].values

# 가중치에 따른 데이터 변형
weighted_data = data * weights.reshape(-1, 1)

# 이너셔 값을 저장할 리스트 생성
inertia_values = []

# 클러스터 수에 따른 이너셔 계산
for k in range(3, 15):  # 적절한 클러스터 수 범위 설정
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(weighted_data)
    inertia_values.append(kmeans.inertia_)

# 엘보우 메서드 그래프 그리기
plt.plot(range(3, 15), inertia_values, marker='o')
plt.xlabel('클러스터 수')
plt.ylabel('이너셔')
plt.title('Elbow Method')

# 이너셔 값의 변화율 계산
inertia_diff = np.diff(inertia_values)
diff_r = np.diff(inertia_diff) / inertia_diff[1:]
plt.plot(range(4, 14), diff_r, marker='o', linestyle='--', color='r', label='이너셔 변화율')

# 관성 값이 급격히 감소하는 지점 찾기
optimal_k_index = np.where(diff_r == max(diff_r))[0][0] + 3  # 인덱스 보정
plt.axvline(x=optimal_k_index, color='g', linestyle='--', label=f'최적 k: {optimal_k_index}')

plt.legend()
plt.show()

print(f"최적의 k는 {optimal_k_index}입니다.")
```

엘보우 사용시 최적 클러스터 = 12

### 실루엣

```python
from sklearn.metrics import silhouette_samples
import matplotlib.cm as cm
```

```python
# 데이터 전처리
X = final_df[['위도', '경도']]

# 가중치 적용
X['위도'] *= final_df['weight']
X['경도'] *= final_df['weight']

# 최적의 클러스터 갯수 탐색
silhouette_scores = []
for n_clusters in range(3, 15):  # 클러스터 갯수 범위 설정
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(X)
    silhouette_avg = silhouette_score(X, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# 클러스터 수에 따른 실루엣 점수 그래프 그리기
plt.plot(range(3, 15), silhouette_scores, marker='o')
plt.xlabel('클러스터 수')
plt.ylabel('실루엣 점수')
plt.title('Silhouette Score by Cluster Number')
plt.grid(True)  # 그리드 표시

# 최적의 클러스터 수 표시
optimal_k = silhouette_scores.index(max(silhouette_scores)) + 3  # 인덱스 보정
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'Optimal k: {optimal_k}')
plt.legend()

plt.show()

print(f"최적의 k는 {optimal_k}입니다.")
```

실루엣 사용시 최적 클러스터 = 9

```python
# Gap 통계량 계산 함수
def compute_gap_statistics(data, k_max=15, B=10):

    # 최대 클러스터 수 설정
    k_range = range(1, k_max + 1)

    # 실제 데이터로 클러스터링을 수행하여 Wk를 계산
    wk = np.zeros(len(k_range))
    wk_err = np.zeros(len(k_range))

    for i, k in enumerate(k_range):
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(data)
        wk[i] = kmeans.inertia_

        # B번의 시뮬레이션을 통해 Wk*를 계산
        bw = np.zeros(B)
        for j in range(B):
            # 데이터 크기와 차원에 맞게 랜덤 데이터 생성
            rand_data = np.random.random_sample(size=data.shape)
            kmeans_rand = KMeans(n_clusters=k, random_state=42)
            kmeans_rand.fit(rand_data)
            bw[j] = kmeans_rand.inertia_
        wk_err[i] = np.sqrt(np.var(bw, ddof=1))

    # Gap 통계량 계산
    gap = np.log(wk) - np.log(wk - wk_err)

    return gap, wk, wk_err

# 데이터 준비
data = final_df[['위도', '경도', 'weight']].values

# Gap 통계량 계산
gap, wk, wk_err = compute_gap_statistics(data)

# 최적 클러스터 수 찾기
optimal_k = np.argmax(gap) + 1

# 클러스터 수에 따른 Gap Statistics 그래프 그리기
plt.plot(range(1, len(gap) + 1), gap, marker='o')
plt.xlabel('클러스터 수')
plt.ylabel('Gap statistic')
plt.title('Gap Statistic for KMeans clustering')
plt.grid(True)  # 그리드 표시

# 최적의 클러스터 수 표시
plt.axvline(x=optimal_k, color='r', linestyle='--', label=f'Optimal k: {optimal_k}')
plt.legend()

plt.show()

print(f"최적의 k는 {optimal_k}입니다.")
```

### n_clusters  = 8

최종 n_cluster -> elbow, silhouette, gap statistic 의 평균 (9+12+4) / 3 = 8.xxx = 8

```python
# 가중치
weights = final_df['weight'].values

# 가중치 정규화
weights = (weights - np.min(weights)) / (np.max(weights) - np.min(weights))

# 위도, 경도, 가중치를 가진 데이터
data = final_df[['위도', '경도']].values

# 클러스터 개수
n_clusters = 8

# KMeans 모델 초기화
kmeans = KMeans(n_clusters=n_clusters)

# 가중치를 반영하여 KMeans 모델을 학습
kmeans.fit(data, sample_weight=weights)

# 클러스터링 결과
clusters = kmeans.labels_

# 각 클러스터의 중심
cluster_centers = kmeans.cluster_centers_

# 각 데이터 포인트가 속한 클러스터
final_df['cluster'] = clusters
```

### 생태통로 입지선정 시각화

진할수록 우선순위

```python
# 클러스터별 weight의 평균 계산 및 정렬
cluster_weights = final_df.groupby('cluster')['weight'].mean().sort_values(ascending=False)

# 초록색 계열 색상 설정
green_colors = ['#006400', '#008000', '#8DB600', '#808000', '#90EE90', '#F5FFFA', '#00FF7F', '#00FF00']

# folium 지도 객체 생성
m = folium.Map(location=[37.5, 127.5], zoom_start=7)

# 모든 데이터 포인트를 지도에 표시
for index, row in final_df.iterrows():
    # 클러스터의 weight 순서에 따라 색상 선택
    cluster_index = cluster_weights.index.get_loc(row['cluster'])
    color = green_colors[cluster_index]
    
    folium.CircleMarker(
        location=[row['위도'], row['경도']],
        radius=4,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=f"Cluster: {row['cluster']}, Weight: {row['weight']}"
    ).add_to(m)
```
